```python
#drop record with no product
df = df.drop(df[df.stockcode.isin(['B', 'D', 'M', 'S'])].index)

#drop record with negative quantity
df = df.drop(df[df.quantity < 0].index)

#drop record with negative price
df = df.drop(df[df.unitprice < 0].index)

#change unspecified country to united kingdom
df.loc[:,'country'] = df.country.str.replace('Unspecified', 'United Kingdom')

#remove character from invoice number
df.loc[:,'invoiceno'] = df.invoiceno.str.strip('C')
df.loc[:,'invoiceno'] = df.invoiceno.str.strip('A')

#match all stockcode with character into uppercase
df.stockcode = df.stockcode.astype(str)
df.stockcode = df.stockcode.str.upper()

#change invoice no datatype into numeric
df.invoiceno = pd.to_numeric(df['invoiceno'])

#change invoicedate datatype into datetime
df.invoicedate = pd.to_datetime(df['invoicedate'], dayfirst=True)

#add month and year
df['invoicemonth'] = df.invoicedate.dt.month
df['invoiceyear'] = df.invoicedate.dt.year
df['invoiceyear'] = df.invoicedate.dt.day

#add totalprice
df['totalprice'] = df.quantity * df.unitprice
```

```python
#create new dataframe for the dictionary
df2 = df[['stockcode', 'description']].dropna()

#delete some improper description
df2 = df2.drop(df2[df2.description == 'mailout'].index)
                   
#delete duplicates based on the stockcode
df2 = df2.drop_duplicates(subset='stockcode', keep='first')

#create dictionary with description as values and stockcode as key
dfdict = pd.Series(df2.description.values,index=df2.stockcode).to_dict()

#apply the dictionary to the original dataset
df.loc[:,'description'] = df.stockcode.map(dfdict)

#drop the remaining NaN in description and customerid
df = df.dropna()

#reset the index of the data
df = df.reset_index(drop=True)

#the new data info
df.info()
```

```python
#import appropriate library
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

#create new dataframe for pivot table
dfpivot = df[['invoiceno', 'description', 'quantity']]

#create a pivot table to group invoices
def return_one(parameter):
    return 1

#pivot table values will be 1 & 0
pivot = pd.pivot_table(dfpivot, values='quantity', index=['invoiceno'],
                       columns=['description'], aggfunc=return_one, fill_value=0)

#apply apriori for frequent item pattern
freq_itemsets = apriori(pivot, min_support=0.03, use_colnames=True)
print("Frequent Itemsets (sets of items that frequently occur together with min. support of 0.03):")
print(freq_itemsets)
```

```python
#import appropriate library
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

#create new dataframe for pivot table
dfpivot = df[['invoiceno', 'description', 'quantity']]

#create a pivot table to group invoices
def return_true(parameter):
    return True

#pivot table values will be True & False
pivot = pd.pivot_table(dfpivot, values='quantity', index=['invoiceno'],
                       columns=['description'], aggfunc=return_true, fill_value=False)

#apply apriori for frequent item pattern
freq_itemsets = apriori(pivot, min_support=0.03, use_colnames=True)
print("Frequent Itemsets (sets of items that frequently occur together with min. support of 0.03):")
print(freq_itemsets)
```

```python
#lower support have pair of items
freq_itemsets2 = apriori(pivot, min_support=0.028, use_colnames=True)
print("Frequent Itemsets (sets of items that frequently occur together with min. support of 0.028:")
print(freq_itemsets2)
```

```python
#lower pattern support for association rule
freq_itemsets3 = apriori(pivot, min_support=0.02, use_colnames=True)

#association rule with minimum confidence 0.6
assoc_rules = association_rules(freq_itemsets3, metric="confidence", min_threshold=0.6)
print("\nAssociation Rules (based on the frequent itemsets):")
assoc_rules.sort_values(by='confidence', ascending=False)
```

```python
# using the silhouette score
silhouette_scores = []
for k in range(2, 12): 
    kmeans = KMeans(n_clusters=k, n_init = 10) 
    kmeans.fit(pivotclust)
    silhouette_scores.append(silhouette_score(pivotclust, kmeans.labels_))
len(silhouette_scores)
```

```python
#plotting silhouette scores
plt.plot(range(2, 12), silhouette_scores, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.show()
```
